```python
experiencia = [1,3,5,7,9]
Salario_data_scientists = [3434,6436,2343,4656,8999]
Salario_ing_Software = [3344,63446,2343,46456,84599]
```

```python
plt.plot(experiencia,Salario_data_scientists)
plt.plot(experiencia,Salario_ing_Software)
plt.title('Salarios')
plt.xlabel('Experiencia')
plt.ylabel('Ganacias')
plt.legend(['Data Scientis', 'Ing en Software'])
plt.show
plt.show
```

```python
plt.plot(experiencia,Salario_data_scientists, color = 'r', label = 'Data Scientists')
plt.plot(experiencia,Salario_ing_Software, color = 'g', label = 'Ing Software')
plt.title('Salarios')
plt.xlabel('Experiencia')
plt.ylabel('Ganacias')
plt.legend(['Data Scientis', 'Ing en Software'])
plt.show
plt.show
```

```python
plt.plot(experiencia,Salario_data_scientists, color = '#0f0f0f', label = 'Data Scientists',linestyle = '--')
plt.plot(experiencia,Salario_ing_Software, color = 'g', label = 'Ing Software',linestyle = ':')
plt.title('Salarios')
plt.xlabel('Experiencia')
plt.ylabel('Ganacias')
plt.legend(['Data Scientis', 'Ing en Software'])
plt.show
plt.show
```

```python
plt.plot(experiencia,Salario_data_scientists, color = '#0f0f0f', label = 'Data Scientists',linestyle = '--', marker = 'o')
plt.plot(experiencia,Salario_ing_Software, color = 'g', label = 'Ing Software',linestyle = ':', marker = '*')
plt.title('Salarios')
plt.xlabel('Experiencia')
plt.ylabel('Ganacias')
plt.legend(['Data Scientis', 'Ing en Software'])
plt.legend()
plt.grid(color = 'r',linestyle = '-', linewidth = 2)
plt.show
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame({'Date':data_range, 'Precios':precios})
data.head()
```

```python
plt.plot(data['Date'], data['Precios'], label = 'Precios')
plt.show()
```
